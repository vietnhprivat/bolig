```python
import requests
import pandas as pd

# Set the API URL and parameters
api_url = "https://services.datafordeler.dk/BBR/BBRPublic/1/rest/enhed"
params = {
    "username": "ANLVUSSNAP",
    "password": "Essvietcc567!",
    "Format": "JSON",
    "Kommunekode": "0751",
    "pagesize": 100000,  # Specify the maximum page size
}

all_data = []
page = 1

while True:
    # Update the page parameter
    params["page"] = page

    # Fetch data from the API with paging parameters
    response = requests.get(api_url, params=params)

    # Check the status code and proceed with processing the response
    if response.status_code == 200:
        try:
            data = response.json()
            df = pd.DataFrame(data)
            all_data.append(df)

            # If the length of the dataframe is less than the page size, stop the loop
            if len(df) < params["pagesize"]:
                break

            # Increase the page number for the next iteration
            page += 1

        except ValueError as e:
            print(f"Failed to decode JSON for page {page}.")
            break

    else:
        print(f"Failed to fetch data for page {page} from the API. Status code: {response.status_code}")
        break

# Concatenate all the dataframes in the list
final_df = pd.concat(all_data)

# Print the first few rows of the DataFrame and the total length
display(final_df.head())
print("Length:", len(final_df))
```
